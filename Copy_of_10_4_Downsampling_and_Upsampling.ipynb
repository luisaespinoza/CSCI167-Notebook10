```python
# Now let's try max-pooling
def maxpool(x_in):
  # Pad the input so its dimensions are divisible by 2
  pad_height = (2 - x_in.shape[0] % 2) % 2
  pad_width = (2 - x_in.shape[1] % 2) % 2
  x_in_padded = np.pad(x_in, ((0, pad_height), (0, pad_width)), mode='edge')

  x_out = np.zeros(( int(np.floor(x_in_padded.shape[0]/2)), int(np.floor(x_in_padded.shape[1]/2)) ))
  # TODO -- write the maxpool routine
  # Replace this line
  # x_out = x_out
  x_out = np.max(x_in_padded.reshape(x_in_padded.shape[0]//2, 2, x_in_padded.shape[1]//2, 2), axis=(1, 3))

  return x_out
```

```python
# Finally, let's try mean pooling
def meanpool(x_in):
  # Pad the input so its dimensions are divisible by 2
  pad_height = (2 - x_in.shape[0] % 2) % 2
  pad_width = (2 - x_in.shape[1] % 2) % 2
  x_in_padded = np.pad(x_in, ((0, pad_height), (0, pad_width)), mode='edge')

  x_out = np.zeros(( int(np.floor(x_in_padded.shape[0]/2)), int(np.floor(x_in_padded.shape[1]/2)) ))
  # TODO -- write the meanpool routine
  # Replace this line
  # x_out = x_out
  x_out = np.mean(x_in_padded.reshape(x_in_padded.shape[0]//2, 2, x_in_padded.shape[1]//2, 2), axis=(1, 3))

  return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
def bilinear(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  x_in_pad = np.zeros((x_in.shape[0]+1, x_in.shape[1]+1))
  x_in_pad[0:x_in.shape[0],0:x_in.shape[1]] = x_in
  # TODO -- write the duplication routine
  # Replace this line
  # x_out = x_out
  for i in range(x_out.shape[0]):
    for j in range(x_out.shape[1]):
      # Calculate the corresponding coordinates in the input image
      r = i / 2.0
      c = j / 2.0

      # Get the integer coordinates of the top-left pixel in the input image
      r0 = int(np.floor(r))
      c0 = int(np.floor(c))
      r1 = int(np.ceil(r))
      c1 = int(np.ceil(c))

      # Calculate the weights for interpolation
      wr = r - r0
      wc = c - c0

      # Perform bilinear interpolation
      # Ensure indices are within the padded input array bounds
      val00 = x_in_pad[r0, c0] if r0 < x_in_pad.shape[0] and c0 < x_in_pad.shape[1] else 0
      val01 = x_in_pad[r0, c1] if r0 < x_in_pad.shape[0] and c1 < x_in_pad.shape[1] else 0
      val10 = x_in_pad[r1, c0] if r1 < x_in_pad.shape[0] and c0 < x_in_pad.shape[1] else 0
      val11 = x_in_pad[r1, c1] if r1 < x_in_pad.shape[0] and c1 < x_in_pad.shape[1] else 0

      x_out[i, j] = (1 - wr) * (1 - wc) * val00 + (1 - wr) * wc * val01 + wr * (1 - wc) * val10 + wr * wc * val11

  return x_out
```
